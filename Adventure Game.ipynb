### Text Based Choose-Your-Own-Adventure Game

```python
def intro():
    """This function will print introduction and acquire starting input"""
    
    print("Hello adventurer, as you know the evil dragon from atop the hill has been\n" + 
    "terrorizing our village for generations, and we have been completely\n" +
    "helpless. That is until now. We have been observing\nyou from a young age " +
    "and see that you are the only one with the potential\nto finally defeat the " +
    "dragon and bring peace to the village.\n")
    
    decision1 = input("The fight will be dangerous and there is no guarantee\nof " +
    "success, but you are our only hope. Will you help us (yes/no): ")
    while decision1 == "yes" or "no":
        return decision1
    else:
        print("Invalid answer")
        decision1 = input("The fight will be dangerous and there is no guarantee\nof " +
        "success, but you are our only hope. Will you help us (yes/no): ")
        return decision1
```

```python
def part_1():
    """This Functions asks for the user's weapon choice"""
    
    print("Hooray! Thank you for agreeing to save our village. You must prepare " +
    "for the fight, however.\nHead to the blacksmith to pick the gear you wish " +
    "to use.\n")
    print("*You go to the blacksmith*\n")
    print("Blacksmith: I hear you are the adventurer who has agreed to fight the dragon.\n" + 
    "I am more than happy to aid you with the gear you need.\n")
    
    print("1. Sword\n2. Bow and Arrows\n3. Magic Wand\n")
    
    weapon = int(input("Which weapon would you like to use in battle? (enter a number): "))
    while weapon not in [1, 2, 3]:
        print("Invalid weapon choice")
        weapon = int(input("Which weapon would you like to use in battle? (enter a number): "))
    if weapon in [1, 2, 3]:
        return weapon
```

```python
def sword():
    """This function asks for the user's plan of action when using a sword"""
    
    print("\n*You go to the evil dragon's hill and notice it is currently asleep*\n")
    print("1. Attack it with your sword\n2. Throw a rock to wake it up\n3. Throw the sword at it")
    
    action = int(input("What is your plan of action (enter a number): "))
    while action not in [1, 2, 3]:
        print("Not valid input, try again.\n")
        action = int(input("What is your plan of action (enter a number): "))
        
    if action == 1: 
        print("\n*You attack the dragon in its sleep but it wakes up. You dodge its first\n" +
        "attack, close your eyes, and slash with your sword. When you open\n" + 
        "your eyes you see the dragon’s head at your feet*")
        return True
              
    elif action == 2:
        print("\n*You throw a rock at the dragon to wake it up, hitting it on the head and " +
        "angering it.\nThe dragon charges at you and before you can react, " +
        "it eats you in one bite*")
        return False
              
    elif action == 3:
        print("\n*You throw your sword at the dragon, and to your surprise it lands right in " +
        "its eye.\nThis is a secret weak spot of the dragon and it dies instantly*")
        return True
```

```python
def bow():
    """This function asks for the user's plan of action when using a bow and arrow"""
    
    print("\n*You go to the evil dragon's hill*\n" + "*You notice that the dragon is currently asleep*")
    
    print("1. Aim for the dragon's head with your bow\n" + "2. Aim for the dragon's body with your bow\n"
          + "3. Sneak up to the dragon and stab it with an arrow")
    
    action = int(input("What is your plan of action (enter a number): "))
    while action not in [1, 2, 3]:
        print("Not valid input, try again.\n")
        action = int(input("What is your plan of action (enter a number): "))
        
    if action == 1:
        print("\n*You shoot an arrow aiming for the dragon’s head and hit one of its\n" +
        "eyes. The eyes of the dragon is its weak spot and it dies instantly*")
        return True
    
    elif action == 2:
        print("\n*You shoot an arrow aiming for the dragon’s body but the arrow " +
        "bounces right off waking the dragon.\nThe dragon charges at you and eats you in one bite*")
        return False
    
    elif action == 3:
        print("\n*You attempt to sneak up on the dragon but step on a stick as you\n" +
        "are right behind it. It wakes up and eats you in one bite*")
        return False
```

```python
def end_part():
    """This function continues the story if the action functions return True"""
    
    print("\n*After killing the dragon you catch a glimpse of\na golden chest under "
    + "a tree near where the\ndragon was sleeping. You go to the chest and\n" +
    "find an abundance of gold*\n")
    print("1. Keep it for yourself\n2. Give it to the village people\n")
    
    action = int(input("What will you do with the riches (enter a number): "))
    while action not in [1, 2]:
        print("Invalid choice")
        action = int(input("What will you do with the riches (enter a number): "))
        
    if action == 1:
        print("\n*You decide go keep the gold for yourself and\ngo far away from the " +
        "village. Two days on the road, however, a bandit sees you\ndrop a " +
        "gold coin and robs you of all the gold.\nYou now have nothing. The end.*")
        
    if action == 2:
        print("\n*You bring the gold back to the village and get a hero's welcome." +
        "\nThe village people elect you as their new king and you rule the\n" +
        "village happily for the rest of your life. The end.*")
        
```

```python
def main():
    """This function combines all the above functions into one using a while loop and nested if loops"""
    
    Continue = "y"
    while Continue == "y" or Continue == "Y":
        decision = intro()
        if decision == "yes":
            weapon = part_1()
            if weapon == 1:
                result = sword()
                if result == True:
                    end_part()
            elif weapon == 2:
                result = bow()
                if result == True:
                    end_part() 
                    
            else:
                print("\n*You go to the evil dragon’s hill and prepare to\n" +
                "fight but soon realize you have no idea how to " +
                "use magic.\nThe dragon eats you in one bite.*")
        else:
            print("That is too bad. I am sorry for wasting your time. Goodbye")
            
        print("*********************************\n")
        
        Continue = input("\nWould you like to play again? y/n: ")
        if Continue == "n" or Continue == "N":
            print("Thank you, Goodbye")
        elif Continue == "y" or Continue == "Y":
            print()
        else:
            print("Invalid Input")
            Continue = input("\nWould you like to play again? y/n: ")
main() # Calls main to run the program
        
```
